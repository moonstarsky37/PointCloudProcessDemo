# 信義路場域

```python
import os, sys, glob, copy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pdal, pyproj, laspy
from tqdm.notebook import tqdm
from numba import jit, njit, prange
from glob import glob
import CSF 
csf = CSF.CSF()
```

```python
%%time
las_folder = r"P:/HDmap/Close_Field/LiDAR/RS2003_Xinyi/remove_outlier_LAS"
fn_list = glob(las_folder+"/*.las")
fn_list = [os.path.basename(i) for i in fn_list]
for fn in fn_list:
    print(fn, end=" ")
    inFile = laspy.file.File(os.path.join(las_folder, fn), mode='r') # read a las file
    points = inFile.points
    xyz = np.vstack((inFile.x, inFile.y, inFile.z)).transpose() # extract x, y, z and put into a list

    csf = CSF.CSF()

    # prameter settings
    csf.params.bSloopSmooth = False
    csf.params.cloth_resolution = 0.5
    # more details about parameter: http://ramm.bnu.edu.cn/projects/CSF/download/

    csf.setPointCloud(xyz)
    ground = CSF.VecInt()  # a list to indicate the index of ground points after calculation
    non_ground = CSF.VecInt() # a list to indicate the index of non-ground points after calculation
    csf.do_filtering(ground, non_ground) # do actual filtering.
    print("Filtering Done!")
    outFile = laspy.file.File(r"P:/HDmap/Close_Field/LiDAR/RS2003_Xinyi/step001_ground_filter/ground/"+fn, 
                              mode='w', header=inFile.header)
    outFile.points = points[ground] # extract ground points, and save it to a las file.
    outFile.close()
    outFile = laspy.file.File(r"P:/HDmap/Close_Field/LiDAR/RS2003_Xinyi/step001_ground_filter/non_ground/"+fn,
                              mode='w', header=inFile.header)
    outFile.points = points[non_ground] # extract ground points, and save it to a las file.
    outFile.close()
    inFile.close()
```

```python
import os, sys, glob, copy
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pdal, pyproj, laspy
from tqdm.notebook import tqdm
from numba import jit, njit, prange
import CSF 
csf = CSF.CSF()
```

```python
las_folder = r"P:/HDmap/Open_Field/LiDAR/QinPu/original/"
fn_list = glob.glob(las_folder+"/*.las")
fn_list_tmp = []
for i in fn_list:
    if not "ground" in i: fn_list_tmp.append(i)
fn_list = [os.path.basename(i) for i in fn_list_tmp]
fn_list
```

```python
las_folder = r"P:/HDmap/Open_Field/LiDAR/QinPu/original/"
fn_list = glob.glob(las_folder+"/*.las")
fn_list_tmp = []
for i in fn_list:
    if not "ground" in i: fn_list_tmp.append(i)
fn_list = [os.path.basename(i) for i in fn_list_tmp]
groundFolder = f"P:/HDmap/Open_Field/LiDAR/QinPu/step001_ground_filter/ground/"
nonGroundFolder = f"P:/HDmap/Open_Field/LiDAR/QinPu/step001_ground_filter/non_ground/"
for fn in tqdm(fn_list):
    print(fn, end=" ")
    if not os.path.exists(groundFolder + fn):
        inFile = laspy.file.File(os.path.join(las_folder, fn), mode='r') # read a las file
        points = inFile.points
         # extract x, y, z and put into a list
        xyz = np.vstack((inFile.x, inFile.y, inFile.z)).transpose().astype(float)
        csf = CSF.CSF()
        # prameter settings
        csf.params.bSloopSmooth = False
        csf.params.cloth_resolution = 0.5
        csf.setPointCloud(xyz)
        ground = CSF.VecInt()  # a list to indicate the index of ground points after calculation
        non_ground = CSF.VecInt() # a list to indicate the index of non-ground points after calculation
        csf.do_filtering(ground, non_ground) # do actual filtering.
        print("Filtering Done!")
        outFile = laspy.file.File(groundFolder + fn, mode='w', header=inFile.header)
        outFile.points = points[ground] # extract ground points, and save it to a las file.
        outFile.close()
        outFile = laspy.file.File(nonGroundFolder + fn, mode='w', header=inFile.header)
        outFile.points = points[non_ground] # extract ground points, and save it to a las file.
        outFile.close()
        inFile.close()
    else:
        print("Filtering Done!")
```
